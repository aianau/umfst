```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
import tensorflow as tf

points = tf.constant(np.random.uniform(0, 1000000, (100, 2)))

plt.scatter(points[:, 0], points[:, 1])
plt.show()
```

```python
data = list(points)
linkage_data = linkage(data, method='single', metric='cityblock')
dendrogram(linkage_data)
plt.show()
```

```python
hierarchical_cluster = AgglomerativeClustering(n_clusters=int(15),affinity='manhattan', linkage='single')
labels = hierarchical_cluster.fit_predict(data)
print(labels)
```

```python
plt.scatter(points[:, 0], points[:, 1], c=labels)
for i in range(points.shape[0]):
    plt.text(points[i, 0]+0.2, points[i, 1]+0.1, i)
plt.show()
```
